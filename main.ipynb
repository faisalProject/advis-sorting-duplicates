```python
import pandas as pd

transkip_nilai = pd.read_csv('dataset/ms_lulusan.csv')

df_transkip_nilai = pd.DataFrame(transkip_nilai)
df_transkip_nilai
```

```python
df_transkip_nilai_sorted = df_transkip_nilai.sort_values(by='tahun_lahir')
df_transkip_nilai_sorted
```

```python
duplikat = df_transkip_nilai.duplicated(subset=['tanggal_lulus'])
duplikat
```

```python
df_cleaned = df_transkip_nilai.drop_duplicates(subset=['nim'], keep=False )
print(df_cleaned)
```

```python
# Mengganti nilai NaN dengan nilai 0
df_filled = df_transkip_nilai.fillna(0)
print(df_filled)
```

```python
# Menghapus data duplikat dan mempertahankan entri pertama
df_keep_first = df_transkip_nilai.drop_duplicates(keep='first')
print("\nDataFrame setelah Menghapus Duplikat (Keep First):")
print(df_keep_first)

# Menghapus data duplikat dan mempertahankan entri terakhir
df_keep_last = df_transkip_nilai.drop_duplicates(keep='last')
print("\nDataFrame setelah Menghapus Duplikat (Keep Last):")
print(df_keep_last)

# Menghapus data duplikat dan menghapus semua entri duplikat
df_keep_false = df_transkip_nilai.drop_duplicates(keep=False)
print("\nDataFrame setelah Menghapus Duplikat (Keep False):")
print(df_keep_false)
```
